### **Convención para escribir código**

Para asegurar un buen entendimiento y práctica de programación, es importante entender que para escribir variables, constantes, funciones y clases, existen distintas convenciones en Python. Esto quiere decir que es un acuerdo y no es algo 'obligatorio'.

#### **Snake Case**

Para escribir nombres de **variables**, **funciones** y **métodos** se utiliza la convención `snake_case`.

Para escribir **variables** se realiza lo siguiente:

````python
mi_nombre = "Gabriel Olmos"

print(mi_nombre)
````

Para escribir **funciones** se realiza lo siguiente:

````python
def mi_primera_funcion():
    pass
````

Para escribir **métodos** se realiza lo siguiente:

````python
class ClasePersona():
    def __init__(self, nombre_persona):
        self.nombre_persona = nombre_persona

    # Esto es un método
    def mostrar_nombre(self):
        print(self.nombre_persona)
````

````python
persona_gabriel = ClasePersona("Gabriel Olmos")
persona_gabriel.mostrar_nombre()
````

No debe escribirse de la siguiente forma el nombre de las variables:

```python

```

#### **Pascal Case**

Para escribir el nombre de las **clases de Python** se utiliza la convención ``PascalCase``, la cual combina palabras poniéndolas todas con la primera letra en mayúscula.

````python
# Esta es una clase en python utilizando la convención.
class ClasePersona():
    def __init__(self, nombre_persona):
        self.nombre_persona = nombre_persona

    # Esto es un método
    def mostrar_nombre(self):
        print(self.nombre_persona)
````

#### **1.3 Camel Case**

La convención **Camel Case** es utilizada en el lenguaje **Javascripts** donde las variables se escriben con la primera letra de la palabra en minúscula y la segunda letra de la palabra con mayúscula.

````python
miNombre = "Gabriel Olmos"
print(miNombre)
````

## **3. Estructura de control**

El control de flujo en Python se refiere a la manera en que el intérprete de Python ejecuta las instrucciones en un programa. El control de flujo en Python se puede lograr mediante diversas estructuras, como las que se muestran en los siguientes apartados.

### **3.1 Condicionales (if, elif, else)**

El siguiente código de Python explica el funcionamiento de los condicionales.

```python
if condicion_1:
    # Bloque de código si la condición_1 es verdadera
elif condicion_2:
    # Bloque de código si la condicion_1 es falsa y la condicion_2 es verdadera
else:
    # Bloque de código si ninguna de las condiciones anteriores es verdadera
```

**Nota: Tiene que todo estar identado para que funcione correctamente el código.**

Debes cambiar los valores de edad.

````python
edad = 8

if edad >= 18:
    print("Eres mayor de edad")
elif edad >= 13:
    print("Eres adolescente")
else:
    print("Eres menor de edad")
````

#### **5.2.2 Bucle while**

El siguiente while será un bucle infinito, no existe ninguna condición para que se detenga.

````python
while True:
    user = input("$")
    print(f"Ingresaste: {user}")
````

Podemos agregar un **break** para que se detenga el while.

````python
while True:
    user = input("$")
    print(f"Ingresaste: {user}")

    if user == "exit":
        break
````

Por otro lado, podemos agregar una condición y cuando sea False se detendrá el while

````python
contador = 1

while contador <= 5: # Acá está la condición, es siempre True hasta que contador sea 6 y al ser False se detiene.
    print(contador)
    contador += 1 # Esto es lo mismo que realizar contador = contador + 1
````

## **4. Programación Orientada a Objetos (OOP)**

La Programación Orientada a Objetos (POO) en Python es un paradigma que permite modelar problemas de manera estructurada y modular. A continuación, se presentan los conceptos básicos de la POO en Python:

### **Clases y Objetos**:

* **Clase**: Es una plantilla o molde que define las propiedades y comportamientos de un objeto. Se define usando la palabra clave **class**.
* **Objeto**: Es una instancia de una clase. Cada objeto tiene sus propias propiedades (atributos) y puede realizar acciones (métodos).

**Ejemplo**: Podemos hacer la analogía con un coche, el cual el coche tiene una estructura metálica que da la forma al coche (Clase). Un objeto sería utilizar esa plantilla para construir un coche con cierto color, ruedas, asientos, etc. Posterior a esto, los objetos de python creados tienen métodos y atributos.
* Los **métodos** es equivalente a decir funciones de python y son las **acciones** que puede hacer el objeto. Siguiendo el ejemplo del coche, los métodos del coche sería moverse, frenar, encender las luces, etc.
* Los **atributos** son las características "estáticas" que tiene ese objeto. Siguiendo el ejemplo del coche, los atributos puede ser el color del coche, cantidad de puertas, cantidad de ruedas, etc. Es importante recalcar que en las clases de python existen los **atributos de clase** y los **atributos de instancias**.

````python
class Coche:
    pass

mi_coche = Coche()  # Objeto mi_coche de la clase Coche
````

### **Métodos**

Son funciones definidas dentro de una clase. Se utilizan para definir las acciones que puede realizar un objeto.

````python
class Coche:
    def arrancar(self):
        print(f"El coche a arrancado.")

mi_coche = Coche()
````

````python
mi_coche.arrancar()  # Salida: El coche a arrancado.
````

### **Atributos**

Son datos asociados a una clase o a una instancia de la clase. Pueden ser:

* Atributos de clase: Compartidos por todos los objetos de la clase que se instancien. Dicho de otro modo, cada objeto declarado tendrán el mismo atributo de clase.
* Atributos de instancia: Únicos para cada objeto. Dicho de otro modo, cada vez que instanciamos un objeto este recibirá parámetros que solicita la función \_\_init_\_, la cual se conoce como **constructor**.

**Nota**: Para acceder a los atributos y métodos debemos utilizar la notación del punto, la cual aparece más adelante.

````python
class Coche:
    color = "Negro"  # Atributo de clase

    def __init__(self, marca, modelo):
        self.marca = marca  # Atributo de instancia
        self.modelo = modelo  # Atributo de instancia

    def arrancar(self):
        return f"El coche a arrancado V2"

mi_coche = Coche("Toyota", "Corolla")
````

````python
# Muestre el atributo marca y modelo
mi_coche.color
````

````python
# Muestre qué hace el método arrancar
mi_coche.arrancar()
````

* **Nota 1**: Para diferenciar entre un atributo y un método, podemos prestar atención en que los atributos no finalizan con los paréntesis.
* **Nota 2**: Los métodos pueden recibir parámetros de entrada. Recuerda que es el equivalente a indicar una función de python.
