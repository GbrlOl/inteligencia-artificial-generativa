#### **1.3.2 Multiplicación Matricial**

Para empezar debemos recordar la multiplicación matricial.

Dadas dos matrices $A$ y $B$, su producto $ C = A \times B $ es posible si el número de columnas de $ A $ es igual al número de filas de $ B $.  

Si $ A $ es de dimensión $ m \times n $ y $ B $ es de dimensión $ n \times p $, entonces $ C $ tendrá dimensión $ m \times p $.  

Cada elemento $ C_{ij} $ de la matriz resultante se obtiene como la suma del producto de los elementos de la fila $ i $ de $ A $ por los elementos de la columna $ j $ de $ B $:

$$
C_{ij} = \sum_{k=1}^{n} A_{ik} B_{kj}
$$

Ejemplo:  

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

Multiplicamos filas por columnas:

$$
C = \begin{bmatrix} (1\times5 + 2\times7) & (1\times6 + 2\times8) \\ (3\times5 + 4\times7) & (3\times6 + 4\times8) \end{bmatrix}
$$

$$
C = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}
$$

```python
a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

b = [
    [1, 2],
    [3, 4],
    [5, 6]
]

matrix_a = torch.tensor(a)
matrix_b = torch.tensor(b)

c = matrix_a @ matrix_b
```

Esta operación matricial se está realizando en la CPU.

#### **1.4.1 Reshape**

Es común tener que cambiar la forma de los tensores, dado que se espera de cierta forma un tensor y se debe convertir a otra dimensión.

El método `reshape()` cambia la forma (dimensiones) de un tensor sin cambiar sus datos. Si los nuevos tamaños no son compatibles con el número de elementos en el tensor original, lanzará un error. A diferencia de view(), reshape() puede realizar copias de los datos si es necesario.

```python
x = torch.linspace(1, 10, 10) # Creamos un tensor que contenga 10 elementos
x
```

```python
x.shape
```

```python
matrix = [list(range(1,10))]
matrix
z = torch.tensor(matrix, dtype=torch.float32)
z
```

```python
z.dtype
```

Podemos apreciar `x` es un tensor 1D y  `z` es un tensor 2D. Podemos verificar lo siguiente:

* x tiene la forma (10, )
* z tiene la forma (1, 9). 

Si hago una multiplicación matricial, ¿Qué ocurre?.

```python
# Intentaremos hacer la multiplicación matricial con matmul
mm = torch.matmul(x, z)
```

```python
# Intentaremos hacer la multiplicación matricial con mm
mm = torch.mm(x, z)
```

El problema es la forma de estos tensores, para solucionar este problema debemos hacer un `reshape`.

* x tiene la forma (10,) y lo convertiremos a la forma (10,1), visto de otro modo, convertiremos un vector de diez elementos a una matriz con diez filas y una columna.

```python
x, x.shape
```

```python
reshape_x = x.reshape(10, 1)
reshape_x
```

```python
# verifica la forma
reshape_x.shape
```

```python
reshape_x.dtype
```

```python
z.dtype
```

```python
# Ahora multipliquemos
mm = torch.matmul(reshape_x, z)
mm
```

Como puedes apreciar, nos entregó un error que nos indica que no podemos realizar la multiplicación dado a que tienen distinto tipo de datos.

```python
# verifica el tipo de dato
```

```python
# reshape_x.dtype
```

```python
# z.dtype
```

Vamos a convertir `z` a un float32 y debemos ahora obtener una matriz 10x9, es decir, 90 elementos.

```python
mm = torch.mm(reshape_x, z.type(torch.float32))
mm
```

```python
mm.numel()
```

#### **1.4.2 Stack**

El método ``stack()`` apila (concatena) una secuencia de tensores a lo largo de una nueva dimensión. Todos los tensores deben tener las mismas dimensiones, excepto en la dimensión en la que se realiza la concatenación.

```python
torch.manual_seed(5)

random_tensor = torch.rand(size=(1,3), dtype=torch.bfloat16)
random_tensor
```

```python
random_tensor.ndim
```

Ahora vamos a concatenar el tensor aleatorio de 2x3 hacia abajo, utilizando dim=0. Esto nos creará 3 muestras o ejemplos del mismo tensor, obteniendo la forma (3, 1, 3)

```python
x_stacked = torch.stack([random_tensor, random_tensor, random_tensor], dim=0)
x_stacked
```

```python
# Muestra el resultado
x_stacked
```

```python
x_stacked.shape
```

Ahora vamos a concatenar el tensor aleatorio de 2x3 en una misma matriz, utilizando dim=1. Esto nos dará como resultado una forma de (1, 3, 3)

```python
y_stacked = torch.stack([random_tensor, random_tensor, random_tensor], dim=1)
y_stacked
```

```python
y_stacked.shape
```

```python
# Muestra el resultado
y_stacked
```

#### **1.4.3 Squeeze**

Elimina las dimensiones de tamaño 1 de un tensor. Por ejemplo, si tienes un tensor de forma ``(1, 3, 1, 5)``, usar ``squeeze()`` eliminará las dimensiones que tienen un tamaño de 1, quedando un tensor de forma ``(3, 5)``.

```python
torch.manual_seed(5)

random_tensor = torch.rand(size=(1,3), dtype=torch.bfloat16)

print("Vemos la dimensión del tensor:", random_tensor.ndim)
print("Vemos la forma del tensor:", random_tensor.shape)
```

```python
random_tensor
```

```python
y_stacked = torch.stack([random_tensor, random_tensor, random_tensor], dim=1)

print("Vemos la dimensión del tensor:", y_stacked.ndim)
print("Vemos la forma del tensor:", y_stacked.shape)
```

```python
y_stacked
```

```python
y_stacked.ndim
```

Podemos apreciar que se agregó una dimensión extra, también conocida como la dimensión del batch. ¿Podemos eliminar esa dimensión?, la respuesta es sí, podemos eliminar la dimensión del batch.

```python
y_stacked_squeeze = y_stacked.squeeze()
```

```python
# Muestra cómo quedó el tensor y visualiza la forma
y_stacked_squeeze.ndim
```

Si podemos eliminar una dimensión, ¿Podemos volver agregar la dimensión?

```python
y_stacked_unsqueeze = y_stacked_squeeze.unsqueeze(dim=0)
```

```python
# Muestra el resultado y juega utilizando dim=1, dim=2
y_stacked_unsqueeze.ndim
```

#### **1.4.4 Permutar**

El método `permute()` reorganiza las dimensiones de un tensor. Este método permite cambiar el orden de las dimensiones del tensor, especificando una nueva secuencia de índices.

```python
x_original = torch.rand(size=(224, 28, 28))

# Permutar nos permite mover los ejes
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Shape del tensor original: {x_original.shape}")
print(f"Shape del tensor permutado: {x_permuted.shape}")
```
